# Combining geocodes using codes already gathered by DataKind

Combined the data from the 01_raw_data_exploration script with the geoid data already gathered by DataKind (because the US Census Geocoder API was going to take about 100 hours to provide the geocode information). 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
pd.set_option('display.max_columns',500)
sns.set()
```

```python
geoid = pd.read_csv('../data/NFIRS_2009_2016_Combined_Census_Tract.csv',dtype=str)
```

```python
geoid.head()
```

```python
geoid2 = geoid[['ID','GEOID','MATCH','MATCH_TYPE','TIGER_LINE_ID','TIGER_LINE_SIDE','X','Y']]
```

```python
geoid2.head()
```

```python
nfirs = pd.read_csv('../data/02_interim/nfirs_cleaned.csv')
```

```python
nfirs = pd.concat([nfirs,geoid2],axis=1)
```

```python
cols_to_drop = ['Unnamed: 0','ID']
col_names = {'X':'lon','Y':'lat'}

nfirs = (nfirs.rename(columns=col_names)
         .drop(cols_to_drop,axis=1))

nfirs.columns = nfirs.columns.str.lower()
```

```python
def fix_leading_zeros(string):
    if pd.notnull(string):
        string = string[:-2]
        if len(string) == 10:
            return '0' + string
        else:
            return string
```

```python
nfirs['geoid'] = nfirs['geoid'].apply(fix_leading_zeros)
```

```python
nfirs[['lon','lat']] = nfirs[['lon','lat']].astype(float)
```
