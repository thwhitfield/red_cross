```python
import folium
from folium import plugins
from folium import Popup
from folium.plugins import MarkerCluster
```

```python
nfirs = pd.read_csv('../data/02_interim/03_nfirs_geoids.csv',dtype={'geoid':str})
```

```python
fatl_fires = nfirs[nfirs['oth_death'] > 0]
danger_fires = nfirs[(nfirs['oth_death'] > 0) | (nfirs['oth_inj'] > 0)]
```

```python
fire_heatmap = folium.Map(location = [45, -102], zoom_start=3)
heat_data = all_fires[['lat','lon']].sample(50000).values.tolist()
plugins.HeatMap(heat_data,radius=10).add_to(fire_heatmap)
fire_heatmap
```
